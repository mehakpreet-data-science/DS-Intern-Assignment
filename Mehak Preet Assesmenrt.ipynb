## HANDING MISSING VALUES

```python
df.isnull().sum()
```

```python
df['timestamp']=pd.to_datetime(df['timestamp'],errors='coerce')
```

```python
print(df['timestamp'])
```

```python
for i in df.columns:
    if df[i].dtype=='object' and i!='timestamp':
        df[i]=pd.to_numeric(df[i],errors='coerce')
```

```python
df.dropna(subset='equipment_energy_consumption')
```

```python
df['equipment_energy_consumption'].isnull()
```

```python
df.fillna(df.mean(numeric_only=True),inplace=True)
```

##  HANDLING THE DATE COLUMN

```python
df['houre']=df['timestamp'].dt.hour
```

```python
df['day_of_week']=df['timestamp'].dt.day_of_week
```

```python
x=df.drop(columns=['timestamp','equipment_energy_consumption'])
y=df['equipment_energy_consumption']
```

### PLOTING OUTLIERS

```python
plt.figure(figsize=(20,6))
sns.boxenplot(data=df)
```

```python



df_new=df.copy()


for i in df.select_dtypes(include=np.number).columns:
    Q1 = df[i].quantile(0.15)
    Q3 = df[i].quantile(0.95)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

 
    df_new[i] = df[i].where((df[i] >= lower_bound) & (df[i] <= upper_bound))

df_new.dropna()
```

### SPLITING DATA 

```python
from sklearn.model_selection import train_test_split
x_test,x_train, y_test,y_train=train_test_split(x,y,test_size=0.2,random_state=20)
```

### MAKING LINEAR REGRESSION MODEL

```python
from sklearn.linear_model import LinearRegression
model_lr=LinearRegression()
model_lr.fit(x_train,y_train)
```

```python
predict_lr=model_lr.predict(x_test)
```

### EVALUATING  LINEAR REGRESSION  MODEL

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
R2_lr=r2_score(y_test,predict_lr)
MAE=mean_absolute_error(y_test,predict_lr)
MSE=mean_squared_error(y_test,predict_lr)
RMSE=np.sqrt(MSE)

print(F'THE R2 SCORE OF LINEAR REGRESSION MODEL {R2_lr}')
print(F'THE MEAN ABSOLUTE ERROR OF LINEAR REGRESSION MODEL {MAE}')
print(F'THE MEAN SQUARE ERROR OF LINEAR REGRESSION MODEL {MSE}')
print(F'THE ROOT MEAN SQUARE ERROR OF LINEAR REGRESSION MODEL {RMSE}')
```

# FEATURE SELECTION USING RANDOM FOREST MODEL

```python
from sklearn.ensemble import RandomForestRegressor
model_FI=RandomForestRegressor(n_estimators=40,max_depth=5,min_samples_split=5,min_samples_leaf=4)
model_FI.fit(x_train,y_train)
```

```python
feature_importance=model_FI.feature_importances_
feature_names=x.columns
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=feature_importance,y=feature_names)
plt.plot()
plt.grid(True)
```

# FEATURE SELECTION USING RECURSIVE FEATURE ELIMINATION

```python
from sklearn.feature_selection import RFE
rfe=RFE(estimator=model_FI,n_features_to_select=10)
rfe.fit_transform(x,y)

```

```python
print(x.columns[rfe.support_])
```

```python
print(x.columns[rfe.ranking_])
```

### FEATURE SELECTION USING PERMUATIOM IMPORTANCE

```python
from sklearn.inspection import permutation_importance
FI=permutation_importance(model_FI,x_test,y_test,n_repeats=10,n_jobs=2,random_state=42)
importance_df = pd.DataFrame({
    'feature': x.columns,
    'importance': FI.importances_mean
}).sort_values(by='importance', ascending=False)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=importance_df.head(10),x=importance_df['importance'],y=importance_df['feature'])
plt.plot()
```
